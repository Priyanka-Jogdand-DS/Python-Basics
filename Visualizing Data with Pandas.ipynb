## Pandas Data Visualization Excercises

```python
import pandas as pd
%matplotlib inline

df3 = pd.read_csv('C:\\Users\\Priyanka\\Desktop\\python2\\Time series\\UDEMY_TSA_FINAL\\03-Pandas-Visualization\\df3.csv')
```

```python
print(len(df3))
print(df3.head())
```

#### so df3 has 500 records and 3 columns. The data represents factory productions numbers and reporteed numbers of defects on certain days of the week.

#### 1. Recreate this scatter plot of 'produced' vs 'defective'. Note the color and size of the points. Also note the figure size. 
#### see if you can figure out how to stretch it in a similar fashion 

```python
df3.plot.scatter(x='produced', y='defective', c='red', figsize=(12,3))
```

```python
df3.plot.scatter(x='produced', y='defective', c='red', figsize=(12,3), s=50)
```

#### 2. Create a histogram of the 'produced' column

```python
df3['produced'].plot.hist()
```

#### 2. Create a histogram of'produced', tightening the x-axis and adding lines between bars

```python
df3['produced'].plot.hist(edgecolor = 'k').autoscale(enable=True, axis='both', tight=True)
```

```python
df3['produced'].plot.hist(edgecolor = 'k').autoscale(enable=True, axis='X', tight=True)
```

#### 3. Create a boxplot that shows 'produced' for each 'weekday'. (hint: this is groupby operation)

```python
df3.boxplot()
```

```python
df3.boxplot(by='weekday')
```

```python
df3.boxplot(by='weekday', figsize=(12,3))
```

```python
df3[['weekday', 'produced']].boxplot(by='weekday', figsize=(12,5))
```

#### 5. Create a KDE plot of the 'defective' column

```python
df3['defective'].plot.kde()
```

```python
df3['defective'].plot.kde(ls=':', lw=5) 
```

```python
df3['defective'].plot.kde(ls='--', lw=5) 
```

#### 7. Create a blended area plot of all the columns for just the rows up to 30. (hint: use .loc)

```python
df3.loc[0:30].plot.area(stacked=False, alpha=0.4)
```

```python
ax= df3.loc[0:30].plot.area(stacked=False, alpha=0.4)
ax.legend(loc=0, bbox_to_anchor=(1.3,0.5))
```
