```python
class dog:
    
    def __init__(self, name, age):
        self.name =name
        self.age = age
        
    def bark(self):
        print("bark bark!")
        
    def doginfo(self):
        print(self.name  +  " is "  +  str(self.age)  +  " year(s) ")
        
    def birthday(self):
        self.age += 1
        
    def setBuddy(self, buddy):
        self.buddy = buddy
        buddy.buddy = self

ozzy = dog("Ozzy", 2)
skippy = dog("skippy", 12)
filou = dog("filou", 10)
print(ozzy.name)
print(ozzy.age)
ozzy.bark()
ozzy.doginfo()

skippy.doginfo()
filou.doginfo()

ozzy.birthday()
print(ozzy.age)

ozzy.setBuddy(filou)
print(ozzy.buddy.name)
print(ozzy.buddy.age)

skippy.setBuddy(ozzy)
print(skippy.buddy.name)
print(skippy.buddy.age)

filou.setBuddy(filou)
print(filou.buddy.name)
print(filou.buddy.age)
        
```

```python
# linear regression
from numpy import random
import matplotlib.pyplot as plt
import numpy as np
pagespeeds = np.random.normal(3.0, 10, 20000)
purchaseamount = 100 - (pagespeeds  + np.random.normal(0, 0.1, 20000))*3
plt.scatter(pagespeeds, purchaseamount)
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_error =  stats.linregress(pagespeeds, purchaseamount)
```

```python
import matplotlib.pyplot as plt
def predict(x):
    return slope * x + intercept
fitline = predict(pagespeeds)
plt.scatter(pagespeeds, purchaseamount)
plt.plot(pagespeeds, fitline, c = 'r')
plt.show()
```

```python
#polynomial regression
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)
pagespeeds = np.random.normal(3.0, 1.0, 1000)
```

```python
purchaseamount = np.random.normal(50.0, 10.0, 1000)/pagespeeds
plt.scatter(pagespeeds, purchaseamount)
```

```python
from scipy import stats
x = np.array(pagespeeds)
y = np.array(purchaseamount)
p4 = np.poly1d(np.polyfit(x,y,3))

```

```python
xp = np.linspace(0, 7, 100)
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y, p4(x))
print(r2)
```

```python
#using train/test polynomial regression to prevent overfitting a polynomial regression
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)
pagespeeds = np.random.normal(3.0, 1.0, 100)
purchaseamount = np.random.normal(50.0, 30.0, 100)/pagespeeds
plt.scatter(pagespeeds, purchaseamount)




```

```python
trainX = pagespeeds[:80]
testX = pagespeeds[80:]

trainY = purchaseamount[:80]
testY = purchaseamount[80:]

x = np.array(trainX)
y = np.array(trainY)
p4 = np.poly1d(np.polyfit(x, y,8))
```

```python
import matplotlib.pyplot as plt
xp= np.linspace(0, 7, 100)
axes = plt.axes()
axes.set_xlim([0, 7])
axes.set_ylim([0, 200])
plt.scatter(x,y)
plt.plot(xp, p4(xp), c = 'r')
plt.show()
```

```python
testX = np.array(testX)
testY = np.array(testY)
axes = plt.axes()
axes.set_xlim([0, 7])
axes.set_ylim([0, 200])
plt.scatter(testX, testY)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(testY, p4(testY))
print(r2)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(np.array(trainY), p4(np.array(trainX)))
print(r2)
```
